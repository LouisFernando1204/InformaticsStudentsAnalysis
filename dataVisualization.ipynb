```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/LouisFernando1204/InformaticsStudentsDataset/main/informaticsstudents_dataset")
df.head()
```

```python
# Among the vibrant community of Informatics students at ITS, there exists a notable cohort 
# consistently achieving a GPA above 3.5. This observation suggests that their academic acumen 
# surpasses the conventional threshold, reflecting a culture of excellence within the department.

df_GPA = df['GPA'].value_counts()
plt.figure(figsize=(8,8))
ax = df_GPA.plot(kind = "bar")
ax.bar_label(ax.containers[0])
plt.title("Distributions of GPA of all Informatics Students at ITS", pad=20)
plt.xlabel("GPA")
plt.ylabel("Number of Students")
plt.legend()
plt.show()
```

```python
# Amidst the academic endeavors at Institut Teknologi Sepuluh Nopember (ITS), an unspoken consensus 
# emerges regarding the duration of independent study among Informatics students. It hints that the 
# average hours dedicated to independent study surpasses the conventional expectation of three hours.

df_HoursofIndependtStudy = df['Hours of Independent Study (per day)'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(df_HoursofIndependtStudy, labels=[f"{hours} hours" for hours in df_HoursofIndependtStudy.index], autopct='%1.2f%%')
plt.title("Distributions of Hours of Independent Study of all Informatics Students at ITS")
plt.legend(title="Hours of Independent Study (per day)", loc="upper right", bbox_to_anchor=(1, 0, 0.5, 1))
plt.show()
```

```python
# Within the academic fabric of Institut Teknologi Sepuluh Nopember (ITS), the cohort from 
# the class of 2019 stands as a testament to excellence in Informatics. Their collective academic 
# performance indicates that the average grade of the class of 2019 surpasses that of the class of 2022.

df.boxplot(by='Class Year', column=['Average Grade in Informatics'], grid=False)
plt.title("Distributions of Average Grade in Informatics by Class Year")
plt.xlabel("Class Year")
plt.ylabel("Average Grade in Informatics")
plt.show()
```

```python
# In the diverse landscape of Informatics at Institut Teknologi Sepuluh Nopember (ITS), 
# male and female students exhibit varied engagement in competitions. This suggests that 
# the average number of competitions participated in by male students differs significantly from that of female students.

df_NumberofCompetitions= df.groupby('Gender')['Number of Competitions'].sum()
plt.figure(figsize=(6, 6))
plt.pie(df_NumberofCompetitions, labels=[f"{competitions}" for competitions in df_NumberofCompetitions.index], autopct='%1.2f%%')
plt.title("Distributions of Number of Competitions by Gender")
plt.legend(title="Number of Competitions", loc="upper right", bbox_to_anchor=(1, 0, 0.5, 1))
plt.show()
```

```python
df_InvolvedinCommunity = df['Involved in Community'].value_counts()
plt.figure(figsize=(8,8))
ax = df_InvolvedinCommunity.plot(kind = "bar")
ax.bar_label(ax.containers[0])
plt.title("Distributions of Students who Involed in Community", pad=20)
plt.xlabel("GPA")
plt.ylabel("Number of Students")
plt.legend()
plt.show()
```
